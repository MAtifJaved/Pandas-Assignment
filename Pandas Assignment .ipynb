# Splitting the GenderPop COlumn into Male and Female Column Seperately

```python
x=US_Census.GenderPop.str.split("_", expand=True)
MalePop1=x[0].str.split("M", expand=True)
FemalePop1=x[1].str.split("F", expand=True)
MalePop=MalePop1[0]
FemalePop=FemalePop1[0]
#display(MalePop)
#display(FemalePop)
US_Census['MalePop']=MalePop
US_Census['FemalePop']=FemalePop
Census_data=US_Census.drop(columns=['GenderPop'])
Census_data
```

```python
Census_data['FemalePop']
```

```python
#To fill the FemalePop Column Empty columns
s=Census_data['FemalePop'].loc[23,31]=np.NaN
#s
#Census_data.fillna(0)
Census_data.columns
```

```python
New_Income_Column=Census_data.Income.str.split("$", expand=True).drop(columns=0)
#print(New_Income_Column)
Census_data['New_Income_Column'] = New_Income_Column
#Census_data
Census_data.columns
```

```python
x=Census_data['FemalePop'].fillna(0).astype('float64')
y=Census_data['New_Income_Column'].fillna(0).astype('float64')
plt.scatter(x,y, label="data1", color='b', s=50, marker="*")
plt.xlabel("Female Population")
plt.ylabel("Income")
plt.legend()
plt.title("US Census\n Scatter Graph")
plt.show()
```
